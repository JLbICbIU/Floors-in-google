```python
# f = open("a_example.in","r")
# f = open("b_small.in","r")
# f = open("c_medium.in","r")
# f = open("d_quite_big.in","r")
f = open("e_also_big.in","r")
contents = f.read()
f.close() 
contents = contents.split('\n')
max_slices = contents[0].split(' ')
num_types = int(max_slices[1])
max_slices = int(max_slices[0])
print("maximum number: {}, number of differrent types: {}".format(max_slices, num_types))
content = contents[1].split(' ')
matrix = [[int(content[j]) for j in range(i,-1,-1)] for i in range(num_types-1,-1,-1)]
    
cur_max = 0
cur_max_s = []
i = 0
j = 0
x = 0
s = []

while j < len(matrix[i]):
    tmp_x = x + matrix[i][j]
    s.append((i,j))
#     for debug
#     for coord in s:
#         print('{}+'.format(matrix[coord[0]][coord[1]]), end="")
#     print('={}'.format(tmp_x))

    if s[0][0] == num_types-1:
        break
    
    if tmp_x == max_slices:
        cur_max = tmp_x
        cur_max_s = s
        break

    if tmp_x < max_slices:

        if tmp_x > cur_max:
            cur_max = tmp_x
            cur_max_s = s
            
        if j == len(matrix[i])-1:
            i = s[-2][0]+1
            j = s[-2][1]
            s = s[:-2]
            x = 0
            for coord in s:
                x += matrix[coord[0]][coord[1]]
        else:
            x = tmp_x
            j+=1

    else:
        if j == len(matrix[i])-1:
            i = s[-2][0]+1
            j = s[-2][1]
            s = s[:-2]
            x = 0
            for coord in s:
                x += matrix[coord[0]][coord[1]]
        else:
            i+=1
            s = s[:-1]
    
print('=====================================')
print('MAX {}: {}'.format(cur_max, cur_max_s))
for coord in cur_max_s:
    print('{}+'.format(matrix[coord[0]][coord[1]]), end="")
print('={}'.format(tmp_x))
```
